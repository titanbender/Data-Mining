````python
# creating the data set....
SF_inputs = {}

SF_inputs['Time'] = list(SF_df.Time)
SF_inputs['Category'] = list(SF_df.Category)
SF_inputs['PdDistrict'] = list(SF_df.PdDistrict)


def one_out_of_K(x):
    #x = SF_inputs['Category'] 
    unique_factors = list(set(x))
    tmp_dict = {}
    for ii in range(0,len(unique_factors)):
        tmp_dict[unique_factors[ii]] = ii
    
    return tmp_dict

def find_key(x,seach_val):
    for key, val in x.items():
        if val == seach_val:
            print key

tmp_cat = one_out_of_K(SF_inputs['Category'])
tmp_pdist = one_out_of_K(SF_inputs['PdDistrict'])


## encoding
SF_inputs['Time'] = [float(int(SF_inputs['Time'][ii][:2]) + int(SF_inputs['Time'][ii][3:]) / 60) for ii in range(0,N)]
SF_inputs['Category'] = [tmp_cat[SF_inputs['Category'][ii]] for ii in range(0,N)]
SF_inputs['PdDistrict'] = [tmp_pdist[SF_inputs['PdDistrict'][ii]] for ii in range(0,N)]

SF_inputs['tree_in'] = [[SF_inputs['Time'][ii], SF_inputs['Category'][ii],SF_inputs['PdDistrict'][ii]] for ii in range(0,N)]



````

````python
# spilt into traning and test..
from sklearn.cross_validation import train_test_split

tree_data = {}
tree_data['X_train'], tree_data['X_test'], _, _ = train_test_split(SF_inputs['tree_in'], [None] * N, test_size=0.10, random_state=42)

tree_data['N_obs'] = N
tree_data['N_train'] = len(tree_data['X_train'])
tree_data['N_test'] = len(tree_data['X_test'])


tree_data['train_labels'] = [tree_data['X_train'][ii][2] for ii in range(0,tree_data['N_train'])]
tree_data['test_labels'] = [tree_data['X_test'][ii][2] for ii in range(0,tree_data['N_test'])]

tree_data['train_obs'] = [tree_data['X_train'][ii][:2] for ii in range(0,tree_data['N_train'])]
tree_data['test_obs'] = [tree_data['X_test'][ii][:2] for ii in range(0,tree_data['N_test'])]

print 'Total number of observations:', tree_data['N_obs']
print 'Number of traning observations:', tree_data['N_train']
print 'Number of testing observations:', tree_data['N_test']
````

````python
from sklearn import tree
tree_data['tree'] = tree.DecisionTreeClassifier()

tree_data['tree_fit'] = tree_data['tree'].fit(tree_data['train_obs'], tree_data['train_labels'])
````

````python
# predictions
tree_data['pre'] = list(tree_data['tree_fit'].predict(tree_data['test_obs']))
tree_data['pre_proba'] = list(tree_data['tree_fit'].predict_proba(tree_data['test_obs']))
````

````python
from sklearn.metrics import accuracy_score
tree_data['acc.'] = accuracy_score(tree_data['test_labels'], tree_data['pre']) * 100
print 'Accuracy of:', tree_data['acc.'], '%'
````

### Digression: Decision trees and unbalanced data

An important problem in many data-science problems is _unbalanced data_. We consider a dataset balanced when the categories we care about have about equal size (e.g. if we want to predict the gender of individuals in the general population). When the category size are imbalanced (e.g. if we are looking for people with a rare disease such as _leukemia_ in the general population), many machine learning algorithms can have problems.

> * Train models of $K = 2,\ldots,10$ on the prostitution data.
> * Explore how the total squared error changes as a function of $K$ and identify what you think is the right number of clusers based on the knee-point in the squared error plot.

````python
from sklearn.cluster import KMeans
#create data frame
XX = [[SF_inputs_pro_cl['lon'][ii],SF_inputs_pro_cl['lat'][ii]] for ii in range(0,N)]
# looping
km_data = {}

K_range = range(2,11)
K_range = range(2,20)

for K in K_range:
    print 'K =', K
    km_data[K] = {}
    km_data[K]['fit'] = KMeans(n_clusters = K, random_state=0).fit(XX)
    km_data[K]['c_center'] = {}
    
    geoplotlib.dot(SF_inputs_pro, point_size = 3, color = 'blue')
    for kk in range(0,K):
        km_data[K]['c_center'][kk] = {}
        km_data[K]['c_center'][kk]['lon'] = list(km_data[K]['fit'].cluster_centers_[kk])[0]
        km_data[K]['c_center'][kk]['lat'] = list(km_data[K]['fit'].cluster_centers_[kk])[1]
        #print km_data[K]['c_center'][kk]
        geoplotlib.dot(km_data[K]['c_center'][kk], point_size = 3, color = 'red')
    
    
    km_data[K]['labels'] = list(km_data[K]['fit'].labels_)
    
    km_data[K]['squared'] = []
    for kk in range(0,K):
        tmp_XX = [XX[ii] for ii in range(0,N) if km_data[K]['labels'][ii] == kk ]
        km_data[K]['squared'].append(np.sum([np.sqrt(np.power((tmp_XX[ii][0] - km_data[K]['c_center'][kk]['lon']),2) + 
                                                     np.power((tmp_XX[ii][1] - km_data[K]['c_center'][kk]['lat']),2)) for ii in range(0,len(tmp_XX))]))
    
    km_data[K]['squared'] = np.sum(km_data[K]['squared'])
    print 'Total squared error:', km_data[K]['squared']
    
    bbox = BoundingBox(north = y_max_lat,
                   west = x_max_lon,
                   south = y_min_lat,
                   east = x_min_lon)
    geoplotlib.set_bbox(bbox)
    geoplotlib.tiles_provider('watercolor')
    geoplotlib.inline()
        
````

````python
errors = [km_data[K]['squared'] for K in K_range]
import matplotlib.pyplot as plt
plt.plot(K_range, errors)
plt.xticks(K_range)
plt.xlabel("K")
plt.ylabel("total squared error")
plt.title("Total Error vs. # of Clusters")
plt.show()
````

````python
# Determine the number of dots and their space size for our heat map
num_of_grid_dots = 25

#initialize a dictionary to contain the longtitude and latitude steps for our grid
lon_lat_grid = []
# the linspace method determines the evenly spaced steps over a specified interval
tmp_lon = list(np.linspace(x_min_lon, x_max_lon, num = num_of_grid_dots))
# determining the step size
step_size = abs(tmp_lon[0] - tmp_lon[1])

tmp_lat = []
ii = 0
# create a list of latitudes based upon the previous defined stepsize
while y_min_lat + step_size * ii <= y_max_lat:
    tmp_lat.append(y_min_lat + step_size * ii)
    ii = ii + 1

for ii in range(0,len(tmp_lon)):
    tmp_grid_lon = tmp_lon[ii]           
    for jj in range(0,len(tmp_lat)):
        tmp_grid_lat = tmp_lat[jj] 
        lon_lat_grid.append([tmp_grid_lon,tmp_grid_lat])
````

> * And by the way: The fit only gets better when we add more means - why not keep adding more of them: Explain in your own words why it makes sense to stop around a knee-point.
> * Another way of estimating the right number of clusters in a $K$-means problem is _stability analysis_. The idea is the following
>   - For each $K = 2,\ldots,10$ generate $N = 10$ clusterings based on random 50% of data (or some other fraction of data/bootstrap).
>   - Divide the space up into an e.g. 20 by 20 grid, and give each gridpoint an index $i$ in the range 1-400. 
>   - Now represent each clustering as a vector $\mathbf{c}^K$, which where the $j$th entry $c_j^K$ is given by the number of centroids in that gridpoint. Most of the entries will be zero, and we have that $\sum_{j = 1}^{400} c_j^K = K$.
>   - We now define _stability_ for some value of $K$ as average pairwise similarity of the $N$ clusterings, where the similarity between clustering $i$ and $j$ is the cosine distance between clustering vectors $\mathbf{c}^K(i)$ and $\mathbf{c}^K(j)$.
>   - We now say that the right $K$ maximizes stability.
